```python
import numpy as np
import pydicom
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
import random
import glob
```

```python
data_input_path = "../input/rsna-miccai-brain-tumor-radiogenomic-classification"
df_train = pd.read_csv(os.path.join(data_input_path, "train_labels.csv"))
sns.set()

plt.figure(figsize = (5,5))
# Count the number of images per category
sns.countplot(x = 'MGMT_value', color = '#169DE3',data = df_train)

plt.title('Categories Distribution'.title(),size=22 , color = '#169DE3')
plt.xlabel('MGMT_value',size=17 , color = '#169DE3')
plt.ylabel('Count',size=17 , color = '#169DE3')

plt.show()
```

```python
#Some images cannot display

def load_dicom(path):
    dicom = pydicom.read_file(path)
    data = dicom.pixel_array
    data = data - np.min(data)
    if np.max(data) != 0:
        data = data / np.max(data)
    data = (data * 255).astype(np.uint8)
    return data

plt.figure(figsize=(16, 5))
for idx, i in enumerate(['FLAIR', 'T1w', 'T1wCE', 'T2w'], 1):
    path = '../input/rsna-miccai-brain-tumor-radiogenomic-classification/train/00000/' + i + '/Image-24.dcm'
    data = load_dicom(path)
    plt.subplot(1, 4, idx)
    plt.imshow(data, cmap="gray")
    plt.title(f"{i}", fontsize=16)
    plt.axis("off")

plt.suptitle(f"MGMT_value: 0", fontsize=16)
plt.show()
```
